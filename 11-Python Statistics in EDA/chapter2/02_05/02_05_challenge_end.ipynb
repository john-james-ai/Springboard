```python
import numpy as np
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
tbfilepath = os.path.join(os.path.join(os.getcwd(),'11-Python Statistics in EDA\chapter2\\02_05'),'tb.csv')
finalpath = os.path.join(os.path.join(os.getcwd(),'11-Python Statistics in EDA\chapter2\\02_05'),'final.csv')
```

```python
tb = pd.read_csv(tbfilepath)
tb.head()
```

```python
# Melt the dataframe into long format
meltcols = [col for col in tb.columns if 'm' in col or 'f' in col]

tb2 = pd.melt(tb, id_vars=['country', 'year'], value_vars=meltcols, var_name='sex_age', value_name='cases')
```

```python
# Create sex column
tb2['sex'] = tb2.apply(lambda row: 'Male' if 'm' in row.sex_age else 'Female', axis=1)
tb2
```

```python
# Create age range
tb2['age'] = tb2['sex_age'].str.slice(1)
tb2
```

```python
# Reformat age to make more readable
tb2['age'] = tb2['age'].map({'04': '0-4', '514': '5-14', '1524': '15-24', '2534': '25-34', '3544': '35-44', '4554': '45-54', '5564': '55-64', '65': '65+', 'u': np.nan})
tb2
```

```python
# Drop sex_aage
tb3 = tb2.drop(['sex_age'], axis=1)
tb3 = tb3[tb3['cases'].notna()]
tb3
```

```python
# Aggregate by ses and age, country and year
tb4 = tb3.groupby(by=['country', 'year', 'sex', 'age']).sum()

tb4
```
