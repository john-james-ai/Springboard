# Stationary Models
Loosely speaking, a time-series is stationary if its statistical properties at time t are similar to those at time t+h for all integers h. We can make this idea precise with the following definitions:

- **Mean Function**
For $X_t$ be a time series $E(X_t^2)<\infty$, the mean function is:
$$\mu_x(t)=E(X_t)$$
- **Covariance Function**
	The covariance function of $\{X_t\}$ is:
	$$\gamma x(r,s) = \text{Cov}(X_r,X_s)=E[(X_r-\mu x(r))(X_s-\mu x(s))]$$
	for all integers $r$ and $s$

Then we have the following definitions
>**Stationary Time Series**
A time series is stationary $(X_1,...X_n)$ and $(X_{1+h},...,X_{n+h})$ have the same joint distribution for all integers $h>0$ and $n>0$.

> **Weakly Stationary Time Series**
${X_t}$ is weakly stationary if 
$$\mu_x(t)\text{ is independent of }t,\text{and}$$
$$\gamma_X(t+h,t)\text{ is independent of }t \text{ for each } h$$

#### Autocorrelation for Stationary Models
**Autocovariance**, for **stationary models** is essentially the covariance of the time series on itself with some lag $h$.

> **Autocovariance function**
Let ${X_t}$ be a stationary time series. The autocovariance function (ACVF) of ${X_t}$ at lag $h$ is:
$$\gamma x(h)=\text{Cov}(X_{t+h},X_t).$$

**Autocorrelation** is the correlation of a time-series on itself with some lag $h$.

> **Autocorrelation Function**
The autocorrelation function (ACF) OF $\{X_t\}$ at lag $h$ is:
$$\rho x(h)\equiv \frac{\gamma x(h)}{\gamma x(0)}=\text{Cor}(X_{t+h},X_t)$$

### Sample Autocorrelation 
We typically do not have a model, we have observed data. To compute the dependence within the data, we use the **sample autocorrelation function** (sample ACF).  

We have the following definitions:
> **Sample Mean**
Let $x_1,...,x_n$ be observations of a time series. The sample mean of $x1, . . . , x_n$ is
$$\bar{x} = \frac{1}{n}\displaystyle\sum_{t=1}^n x_t$$

> **Sample Autocovariance Function**
The sample autocovariance function is:
$$\hat{\gamma}(h):=\frac{1}{n}\displaystyle\sum_{t=1}^{n-|h|}(x_{t+|h|}-\bar{x})(x_t-\bar{x}),\space\space -n<h<n.$$

> **Sample Autocorrelation Function**
The sample autocorrelation function is:
$$\hat{\rho}(h)=\frac{\hat{\gamma}(h)}{\hat{\gamma}(0)},\space\space -n<h<n.$$

Sample autocovariance and sample autocorrelation functions can be computed for any data set, not just stationary time series.
